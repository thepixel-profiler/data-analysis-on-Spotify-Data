```python
import pandas as pd #importing pandas library for data cleaning and manipulation
import matplotlib.pyplot as plt #importing matplotlib library to use for visualizations
```

```python
#loading the excel file into the notebook for further use
spotify_data = pd.read_csv("D:/Personal Projects/Spotify Analysis/Dataset/dataset.csv", header=0)\
#printing the first 5 rows of the data to see if it was loaded properly or not
spotify_data.head()
```

```python
#obtaining a summary of the dataframe for better understanding of the data
spotify_data.info()
```

```python
#seeing if there are any null values in the dataframe using isnull()
#adding all the null values in a column if there are any using sum()
spotify_data.isnull().sum()
```

```python
#seeing if there are any duplicate rows in the dataframe using duplicated()
#adding all the duplicate rows in each column using sum()
spotify_data.duplicated().sum()
```

```python
spotify_data.head(4)
```

```python
#storing unnecessary columns in a list for the next step
columns_to_drop = ['danceability', 'loudness', 'speechiness', 'instrumentalness']
#dropping unnecessary columns to make the dataframe easier to work with 
spotify_data.drop(columns_to_drop, axis=1, inplace=True)
```

```python
#printing the dataframe to check if the columns were successfully deleted or not
spotify_data.head()
```

```python
#using describe() to generate descriptive statistics of the dataframe
spotify_data.describe()
```

```python
#creating a new dataframe that contains only numerical columns of the data
spotify_data_numeric = spotify_data.select_dtypes(include=['number'])
```

```python
#printing the newly formed dataframe to check 
spotify_data_numeric.head()
```

```python
#generating the correlation matrix of the numerical dataframe
correlation_matrix = spotify_data_numeric.corr()
```

```python
#the obtained correlation matrix
correlation_matrix
```

```python
#importing seaborn library to make a heatmap of the obtained correlation matrix. 
import seaborn as sns
#making a heatmap of the matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Spotify Song Features')
plt.show()
```

```python
# Grouping the data by genre and calculating the mean popularity for making a bar plot
genre_popularity = spotify_data.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

# Creating a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_popularity.values, y=genre_popularity.index, palette='viridis')
plt.title('Top 10 Genres by Average Popularity')
plt.xlabel('Average Popularity Score')
plt.ylabel('Genre')
plt.show()
```

```python
# Grouping the data by genre and calculating the mean popularity
genre_popularity = spotify_data.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

# Creating another bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_popularity.values, y=genre_popularity.index, palette='viridis')
plt.title('Top 10 Genres by Average Popularity')
plt.xlabel('Average Popularity Score')
plt.ylabel('Artists')
plt.show()
```

```python
#filtering the data that has genre as 'k-pop' from the data
kpop_df = spotify_data[spotify_data['track_genre'] == 'k-pop']

# Grouping the data by 'artist' and calculating the average popularity
kpop_artist_popularity = kpop_df.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

# Creating the bar plot for the top 10 K-pop artists
plt.figure(figsize=(12, 6))
sns.barplot(x=kpop_artist_popularity.values, y=kpop_artist_popularity.index, palette='magma')
plt.title('Top 10 K-pop Artists by Average Popularity')
plt.xlabel('Average Popularity Score')
plt.ylabel('Artists')
plt.show()
```
